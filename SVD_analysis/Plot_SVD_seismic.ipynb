```python
import matplotlib.pyplot as plt
import numpy as np
from typing import Optional
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.optim import Adam
import tqdm
device = torch.device("cuda:0" if torch.cuda.is_available() else "cpu")
#print('Current cuda device: ',torch.cuda.get_device_name(0))
```

```python
import matplotlib.colors as colors
```

```python
import pickle
# f = open('no_encoding_USV_block.pkl', 'rb')
f = open('no_encoding_USV_Case_1.pkl', 'rb')
U = pickle.load(f)
S = pickle.load(f)
V = pickle.load(f)
f.close()
```

```python
plt.scatter(np.arange(0,S.size()[0],1), S, s=2)
plt.ylabel('$\lambda_i$')
plt.xlabel('i')
```

```python
fig, axs = plt.subplots(2,5, figsize=(50,40))
fig.tight_layout()
plt.title('After training (linear scale)')
for k in range(5):
    axs[0, k].imshow(U[:,k].reshape((64,128)).t(),  'RdBu', origin = 'lower', norm=colors.CenteredNorm())
    axs[0, k].tick_params(bottom=False,left=False,labelleft=False,labelbottom=False)
    axs[0, k].set_title('$\lambda_{'+str(k)+'}=$' +str(round(S[k].cpu().item() ,3)) , fontsize=40)
for k in range(5):
    axs[1, k].imshow(U[:,k+5].reshape((64,128)).t(),  'RdBu', origin = 'lower', norm=colors.CenteredNorm())
    axs[1, k].tick_params(bottom=False,left=False,labelleft=False,labelbottom=False)
    axs[1, k].set_title('$\lambda_{'+str(k+5)+'}=$' +str(round(S[k+5].cpu().item() ,3)) , fontsize=40)
    
plt.savefig('After_training_(linear_scale)_Case_1.png')
```
